¿Qué porciento de lugares cuentan con facilidades? con este código damos respuesta a esta interrogante para el conocimiento de la personas ya que es una de las preguntas más mencionadas en las páginas web.La verdad es que a pesar de los avances, un elevado porcentaje de lugares no cuentan con estas facilidades,dificultando la movilidad de personas con discapacidad,adultos mayores,los usuarios de sillas de ruedas,andadores o carritos de bebé y eventos de cualquier índole.
Estos espacios que nos presentan una accesibilidad garantizada para su circulación representa un obstáculo que impide que los clientes disfruten de una estancia agradable en estos lugares.Por lo que es necesario trabajar con compromiso para construir una sociedad más inclusiva y accesible para todos 

```python
# Facilidades

m = [an,by,ch,ct,cr,do,gb,he,hv,ll,mr,py,pr,rg,sm]

list = []
it = []
js = []

for a in m:
    for file in os.listdir(a):
        if file.endswith(".json"):
            rute = os.path.join(a, file)
            with open(rute) as n:
                mini = json.load(n)
                js.append(mini)
total = len(js)
def facilities(f: str):
    l1 = []
    l2 = []
    for a in os.listdir(f):
        if a.endswith(".json"):
            rute = os.path.join(f, a)
            with open(rute) as s:
                ms = json.load(s)
                for b in ms.items():
                    l1.append(b)
    n = len(l1)
    for c in range(11,n,18):
        f = l1[c][1]
        if f != "null":
            l2.append(1)
    return np.count_nonzero(l2)

fan = facilities(an)
fby = facilities(by)
fch = facilities(ch)
fct = facilities(ct)
fcr = facilities(cr)
fdo = facilities(do)
fgb = facilities(gb)
fhe = facilities(he)
fhv = facilities(hv)
fll = facilities(ll)
fmr = facilities(mr)
fpy = facilities(py)
fpr = facilities(pr)
frg = facilities(rg)
fsm = facilities(sm)

f = [fan,fby,fch,fct,fcr,fdo,fgb,fhe,fhv,fll,fmr,fpy,fpr,frg,fsm]
fl = []

for i in f:
    fl.append(i)
part = np.sum(fl)
facility = (part / total) * 100
print(f"El por ciento de lugares que cuentan con facilidades es del {facility} %")

x = ['AN','BY','CH','CR','CT','DO','GB','HE','HV','LL','MR','PY','PR','RG','SM']

plt.bar(x, f, color = ["b","g","r", "c", "m","y","b","g","r", "c", "m","y", "b", "r"])
plt.title("Cantidad de lugares que poseen facilidades")
plt.show()
```

```python
# Places

places = []
loc = []

for y in js:
    for z in y.items():
        loc.append(z)
lnl = len(loc)

for nm in range(0,lnl,18):
    title = loc[nm]
    name = title[1]
    places.append(name)

# coordenadas
lt = []
ltd = []

for k in range(17,lnl,18):
    t = loc[k]
    ld = t[1]
    for v in ld.values():
        lt.append(v)
for l in range(2, len(lt), 3):
    lat = lt[l]
    ltd.append(lat)


lat = []
long = []
type = []
phone0 = []
phone = []
# Latitud

for ll in ltd:
    lat.append(ll[0])
    # Longitud

for lgt in ltd:
    long.append(lgt[1])

# Desciption

for num in range(3,lnl,18):
    t = loc[num]
    type.append(t[1])

# Telefono

for tel in range(1,lnl,18):
    p = loc[tel]
    phone0.append(p[1])
for t in phone0:
    if t.count("null") == 0:
        phone.append(t)
    else:
        phone.append("el local no cuenta con telefono")

```

```python
coordinates =  {
    "name": places,
    "latitud": lat,
    "longitud": long,
    "type": type,
    "phone": phone
}

df = pd.DataFrame(coordinates)
df
```
