```python
value = []
name = []
def kitchen(k: str):
    k1 = []
    k2 = []
    txt = []
    for b in js:
        for c in b.items():
            k1.append(c)
    lk = len(k1)
    for d in range(4,lk,18):
        y = k1[d][1]
        v = y.count(k)
        if v > 0:
            k2.append(v)
            txt.append(str(k))
    for i in range(len(txt)):
        if txt[i] not in name:
            name.append(txt[i])

def median(m: str, d: str):
    list = []
    n = []
    median = []
    for file in os.listdir(m):
        if file.endswith(".json"):
            rute = os.path.join(m, file)
            with open(rute) as js:
                data = json.load(js)
                n.append(data)
    for i in n:
        for o in i.values():
            list.append(o)
    ln = len(list)
    for x in range(16,ln,18):
        b = list[x]
        for z in b:
            if z == d:
                c = b[d]                    
                for y in c.values():
                    for w in y.values():
                        if type(w) == int or type(w) == float:
                            median.append(w)
                        else:
                            for p in w.values():
                                if type(p) == int or type(p) == float:                                        
                                    median.append(p)
                                else:
                                    for t in p.values():
                                        if type(t) == int or type(t) == float:
                                            median.append(t)
                                        else:
                                            for o in t.values():
                                                if type(o) == int or type(o) == float:
                                                    median.append(p)
                                                else:
                                                    for u in o.values():
                                                        median.append(u)                                                                      
                else:
                    continue
    median = np.median(median)          
    return median
```

```python
# Facilidades

m = [an,by,ch,ct,cr,do,gb,he,hv,ll,mr,py,pr,rg,sm]

list = []
it = []
js = []

for a in m:
    for file in os.listdir(a):
        if file.endswith(".json"):
            rute = os.path.join(a, file)
            with open(rute) as n:
                mini = json.load(n)
                js.append(mini)
total = len(js)
def facilities(f: str):
    l1 = []
    l2 = []
    for a in os.listdir(f):
        if a.endswith(".json"):
            rute = os.path.join(f, a)
            with open(rute) as s:
                ms = json.load(s)
                for b in ms.items():
                    l1.append(b)
    n = len(l1)
    for c in range(11,n,18):
        f = l1[c][1]
        if f != "null":
            l2.append(1)
    return np.count_nonzero(l2)

fan = facilities(an)
fby = facilities(by)
fch = facilities(ch)
fct = facilities(ct)
fcr = facilities(cr)
fdo = facilities(do)
fgb = facilities(gb)
fhe = facilities(he)
fhv = facilities(hv)
fll = facilities(ll)
fmr = facilities(mr)
fpy = facilities(py)
fpr = facilities(pr)
frg = facilities(rg)
fsm = facilities(sm)

f = [fan,fby,fch,fct,fcr,fdo,fgb,fhe,fhv,fll,fmr,fpy,fpr,frg,fsm]
fl = []

for i in f:
    fl.append(i)
part = np.sum(fl)
facility = (part / total) * 100
print(f"El por ciento de lugares que cuentan con facilidades es del {facility} %")

x = ['AN','BY','CH','CR','CT','DO','GB','HE','HV','LL','MR','PY','PR','RG','SM']

plt.bar(x, f, color = ["b","g","r", "c", "m","y","b","g","r", "c", "m","y", "b", "r"])
plt.title("Cantidad de lugares que poseen facilidades")
plt.show()
```
